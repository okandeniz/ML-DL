## PCA

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.1,random_state=42)
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
## standardlaştırma:

sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
pca=PCA(n_components=10) # 30 bileşeni 10 bileşene düşürmek istiyoruz..
```

```python
X_train2=pca.fit_transform(X_train)
X_test2=pca.transform(X_test)
```

```python
print(X_train.shape) # 30 değişken
print(X_train2.shape) # 10 değişken
```

En az bileşeni kullanacağız ve hem de en az varyans kaybını kullanmamız gerekiyor:

```python
import numpy as np
# Açıklanan varyans:
print(np.cumsum(pca.explained_variance_ratio_)*100) # %95 açıklanan varyans iyi..
```

```python
plt.plot(np.cumsum(pca.explained_variance_ratio_)*100)
plt.xlabel("Bileşen Sayısı")
plt.ylabel("Açıklanan Varyans")
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split,GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVC,SVC
from sklearn.linear_model import RidgeClassifier,LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from lightgbm import LGBMClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
```

```python
models=["LinearSVC","SVC","Ridge","Logistic","Random Forest","LGBM","XGBM"]
siniflar=[LinearSVC(random_state=0),SVC(random_state=0),RidgeClassifier(random_state=0),
          LogisticRegression(random_state=0),RandomForestClassifier(random_state=0),LGBMClassifier(random_state=0),XGBClassifier()]

parametreler={
    models[0]:{"C":[0.1,1,10,50,100,150],"penalty":["l1","l2"]},
    models[1]:{"kernel":["linear","rbf"],"C":[0.1,1,10,50,100,150],"gamma":[0.1,0.01,0.001]},
    models[2]:{"alpha":[0.1,1.0,1.5,2],"solver":["auto","svd","cholesky","lsqr","sparse_cg","sag","saga","bfgs"]},
    models[3]:{"C":[0.1,1,1.5,2],"penalty":["l1","l2","elasticnet"],"solver":["lbfgs","liblinear","newton-cg","newton-cholesky","sag","saga"]},
    models[4]:{"n_estimators":[100,500,1000,2000],"max_depth":[4,10],"min_samples_split":[2,5]},
    models[5]:{"learning_rate":[0.001,0.01],"n_estimators":[100,500,1000,2000],"max_depth":[4,10],"subsample":[0.6,0.8]},
    models[6]:{"learning_rate":[0.001,0.01],"n_estimators":[100,500,1000,2000],"max_depth":[4,10],"subsample":[0.6,0.8]}
}
```

```python
def cozum(model):
    model.fit(X_train2,y_train)
    return model
```

```python
def skor(model2):
    tahmin=cozum(model2).predict(X_test2)
    asc=accuracy_score(y_test,tahmin)
    return asc*100
```

```python
## hiper parametre tune:
for i,j in zip(models,siniflar):
    print(i)
    grid=GridSearchCV(cozum(j),parametreler[i],cv=10,n_jobs=-1)
    grid.fit(X_train2,y_train)
    print(grid.best_params_)
```

```python
models=["LinearSVC","SVC","Ridge","Logistic","Random Forest","LGBM","XGBM"]
siniflar=[LinearSVC(random_state=0,C=0.1,penalty="l2",dual=False),SVC(random_state=0,C=0.1,gamma=0.1,kernel="linear"),RidgeClassifier(random_state=0,alpha=0.1,solver="auto"),
          LogisticRegression(random_state=0,C=1,penalty="l2",solver="liblinear",dual=False),RandomForestClassifier(random_state=0,max_depth=10,min_samples_split=2,n_estimators=1000),
          LGBMClassifier(random_state=0,learning_rate=0.01,max_depth=10,n_estimators=1000,subsample=0.6),
          XGBClassifier(learning_rate=0.01,max_depth=4,n_estimators=2000,subsample=0.6)]

def cozum(model):
    model.fit(X_train2,y_train)
    return model

def skor(model2):
    tahmin=cozum(model2).predict(X_test2)
    asc=accuracy_score(y_test,tahmin)
    return asc*100
```

```python
basari=[]

for i in siniflar:
    basari.append(skor(i))
    
a=list(zip(models,basari))
sonuc=pd.DataFrame(a,columns=["Model","Başarı"])
print(sonuc.sort_values("Başarı",ascending=False))
```
